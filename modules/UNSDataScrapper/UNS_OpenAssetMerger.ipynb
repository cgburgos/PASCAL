```python
import pandas as pd
import json
```

```python
with open("FullUNSData.json", "r") as jsonFile:
    dataObject = json.load(jsonFile)
```

```python
df = pd.DataFrame(dataObject)
df = df.drop_duplicates(subset="title",keep='first', ignore_index=True)
```

```python
df.to_json("CleanUNSData.json", orient="records")
```

```python
openAssetDF = pd.read_excel("ProjectDB.xlsx", "Sheet4")
```

```python
openAssetNames = list(openAssetDF["Name"])
webNames = list(df["title"])
```

```python
matchingNames = []
for project in openAssetNames:
    if(project in webNames):
        matchingNames.append(project)
print(matchingNames)
```

```python
mergedDF = pd.merge(openAssetDF, df, left_on="Name", right_on="title", how="left")
mergedDF
```

```python
mergedDF.to_json("MergedProjectDB.json", orient="records")
```

```python
def MergeColumns(df, main, secondary):
    mainVal = df[main]
    secondaryVal = df[secondary]
    outVal = mainVal
    if  str(mainVal) == "None" or str(mainVal) == "nan"	or str(mainVal) == "NaN":
        outVal = secondaryVal
    return outVal
```

```python
mergedDF["Year"] = mergedDF.apply(MergeColumns, axis=1, args=("Year", "year"))
mergedDF["UNStudio Team"] = mergedDF.apply(MergeColumns, axis=1, args=("UNStudio Team", "UNS Team"))
mergedDF["city"] = mergedDF.apply(MergeColumns, axis=1, args=("city", "City (Project Keyword Category)"))
mergedDF["country"] = mergedDF.apply(MergeColumns, axis=1, args=("country", "Country (Project Keyword Category)"))
mergedDF["External Team"] = mergedDF.apply(MergeColumns, axis=1, args=("External Team", "Project Advisors"))
mergedDF["client"] = mergedDF.apply(MergeColumns, axis=1, args=("Client_x", "Client_y"))
mergedDF["program"] = mergedDF.apply(MergeColumns, axis=1, args=("Programme_x", "Programme_y"))
mergedDF["Project Size"] = mergedDF.apply(MergeColumns, axis=1, args=("Project Size", "Building surface"))
mergedDF["Site Area"] = mergedDF.apply(MergeColumns, axis=1, args=("Site Area", "Building site"))
mergedDF["Status"] = mergedDF.apply(MergeColumns, axis=1, args=("Status (Project Keyword Category)", "Status_y"))
```
