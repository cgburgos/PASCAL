```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import time
import pandas as pd
```

```python
import json
with open("UNSProjectData.json", "r") as jsonFile:
    loadedJason = json.load(jsonFile)

filteredProjects = []
for i in loadedJason:
    if len(i.keys()) > 1:
        filteredProjects.append(i)
```

```python
df = pd.DataFrame(filteredProjects, columns=["title", "city", "country", "year", "Client", "Status","Programme", 'Building surface',
 'Building site',"link"])
```

```python
df["Building surface"] = df["Building surface"].str.replace('m2', '')
df["Building surface"] = df["Building surface"].str.replace('\D+', '')
df["Building surface"] = pd.to_numeric(df["Building surface"], errors='coerce')
df["Building site"] = df["Building site"].str.replace('m2', '')
df["Building site"] = df["Building site"].str.replace('\D+', '')
df["Building site"] = pd.to_numeric(df["Building site"], errors='coerce')
df
```

```python
df.to_json("CleanUNSData.json", orient="records")
```

```python
projectsLink = df.link.to_list()
projectsLink
```

```python
projectObjects = []
newObjectsArray = []
driver = webdriver.Chrome()
#[96:98]
for link in projectsLink:
    driver.get(link)
    time.sleep(5)
    newObject = {}
    keywords = []
    concepts = []
        
    try:
        projectTitle = driver.find_element(By.CLASS_NAME, "project-information__content").find_element(By.TAG_NAME, "h1").text
        description = driver.find_element(By.CLASS_NAME, "project-information__content").find_element(By.TAG_NAME, "p").text
        print(projectTitle)
        newObject["description"] = description
        newObject["title"] = projectTitle
    except:
        pass
    
    try:
        
        stickyList = driver.find_elements(By.CLASS_NAME, "section-text-content-sticky")
        for stick in stickyList:
            currStickyArray = stick.text.split("\n")
            if len(currStickyArray[0]) < 50:
                keywords.append(currStickyArray[0])
            concepts.append(stick.text)
    
    except:
        pass
    try:
        teamArray = driver.find_element(By.CLASS_NAME, "project-credits-text").find_elements(By.TAG_NAME, "p")
        for part in teamArray:
            teamComponents = part.text.split("\n\n")
            unstudioTeam = teamComponents[0].split("\n")[1:]
            cleanUnstudioTeam = unstudioTeam[0].replace(" and ", ", ")
            cleanUnstudioTeam = unstudioTeam[0].replace(" with ", ", ")
            newObject["UNStudio Team"] = cleanUnstudioTeam.split(", ")
            newObject["External Team"] = teamComponents[1:]
    except:
        pass
    newObject["keywords"] = keywords
    newObject["concepts"] = concepts
    newObjectsArray.append(newObject)
    print("------------------------------------")
    
```

```python
newObjectsArray
```

```python
softDF = pd.DataFrame(newObjectsArray)

fullDF = softDF.merge(df, on="title", how="right")
fullDF
```

```python
fullDF.to_json("FullUNSData.json", orient="records")
```
