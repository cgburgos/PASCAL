```python
import pandas as pd
import json
```

```python
with open("MergedCleanProjectDB.json", "r") as jsonFile:
    dataObject = json.load(jsonFile)
```

```python
ArchArrays  = []
for object in dataObject:
    currVal = object["UNStudio Team"]
    newVal = []
    if type(currVal) == str:
        currVal = currVal.replace(" and ", ", ")
        currVal = currVal.replace("\nand ", ", ")
        currVal = currVal.replace("\xa0", ", ")
        currVal = currVal.replace(" with ", ", ")
        newVal = currVal.split(", ")
    if type(currVal) == list:
        tempVal = []
        for val in currVal:
            val = val.replace(" and ", ", ")
            val = val.replace("\nand ", ", ")
            val = val.replace("\xa0", ", ")
            val = val.replace(" with ", ", ")
            newVal = val.split(", ")
            for temp in newVal:
                tempVal.append(temp)
        newVal = tempVal
    object["UNStudio Team"] = newVal
    ArchArrays.append(newVal)
```

```python
with open("UNSFullDB.json", "w") as jsonFile:
    json.dump(dataObject, jsonFile, indent=2)
```

```python
uniqueArchitects = []
allArchitects = [] 
for project in ArchArrays:
    for arch in project:
        if arch not in uniqueArchitects:
            uniqueArchitects.append(arch)
        allArchitects.append(arch)

print("Historical Architects: ",len(uniqueArchitects))

uniqueArchitectsCount = []
for arch in uniqueArchitects:
    projectCount = allArchitects.count(arch)
    uniqueArchitectsCount.append(projectCount)
```

```python
df = pd.DataFrame(list(zip(uniqueArchitects, uniqueArchitectsCount)), columns =['Architect', 'Project Count'])
df.sort_values(by=['Project Count'], ascending=False).head(50)
```
